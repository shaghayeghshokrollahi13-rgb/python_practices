```python
#5
def f(num1,num2,num3):
    if(num1>=num2)and(num1>=num3):
        largest=num1
    elif (num2>=num1)and(num2>=num3):
        largest=num2
    else:
        largest=num3
    print("The largest number is:",largest)
num1=float(input("Enter firset number:"))
num2=float(input("Enter second number:"))
num3=float(input("Enter thist number:"))
f(num1,num2,num3)
    
```

6.برنامه مقایسه طول دو رشته با استفاده از تابع

```python
#6
def compare_length(stringe1,stringe2):
    str1=len(stringe1)
    str2=len(stringe2)
    if str1>str2:
        print(stringe1,"is less than",stringe2)
    elif str1<str2:
        print(stringe1,"is greater than",stringe2)
    else:
        print(stringe1,stringe2,"are equal")
stringe1=input("First_name:")
stringe2=input("Last_name:")
compare_length(stringe1,stringe2)
```
