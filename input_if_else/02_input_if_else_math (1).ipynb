تبدیل فرمول ریاضی به کد زیر
$$
y=ln  x  if x>=5
$$
$$
y= \ sqrt {x}  if 0 <= x <5
$$

```python
from math import*
x=float(input("x="))
if x>=5:
    y=log(x)
    print("y=log(x)",y)
elif 0<=x:
    if x<5:
        y=sqrt(x)
        print("y=sqrt(x):",y)
```
